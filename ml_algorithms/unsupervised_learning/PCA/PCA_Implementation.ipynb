```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

print(cancer.keys())
```

```python
pd.DataFrame(X)
```

```python
print(cancer.DESCR)  
```

```python
df = pd.DataFrame(X, columns=cancer.feature_names)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

pd.DataFrame(X_scaled, columns=cancer.feature_names).head()
```

```python
from sklearn.decomposition import PCA

pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled)

print("Explained variance ratio:", pca_2.explained_variance_ratio_)
print("Total variance captured:", pca_2.explained_variance_ratio_.sum())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_2[:, 0], X_pca_2[:, 1], c=y, cmap="plasma")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA (2 components)")
plt.show()
```

```python
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_scaled)

print("Number of components for 95% variance:", pca_95.n_components_)
print("Total variance captured:", pca_95.explained_variance_ratio_.sum())
```

```python
pca_full = PCA().fit(X_scaled)
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker="o")
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.title("Explained Variance vs Components")
plt.axhline(0.95, color="r", linestyle="--")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

results = []

for k in [2, 10]:
    pca = PCA(n_components=k)
    X_pca = pca.fit_transform(X_scaled)

    # variance captured
    var = pca.explained_variance_ratio_.sum()

    # reconstruction error
    X_recon = pca.inverse_transform(X_pca)
    recon_mse = np.mean((X_scaled - X_recon) ** 2)

    # classification accuracy
    clf = LogisticRegression(max_iter=2000)
    acc = cross_val_score(clf, X_pca, y, cv=5, scoring="accuracy").mean()

    results.append([k, var, recon_mse, acc])

pd.DataFrame(results, columns=["n_components", "variance_captured", "recon_mse", "cv_accuracy"])
```
